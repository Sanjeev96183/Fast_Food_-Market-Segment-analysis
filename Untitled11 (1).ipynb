```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

d1 = pd.read_csv('/content/mcdonalds.csv')
d1
```

```python
d1 = pd.DataFrame(d1)
mdx = d1.iloc[:, :11].values # converting the first 11 coloum as matrix #
mdx = (mdx == "Yes").astype(int)
mean = np.round(mdx.mean(axis=0),2) # mean has been calculated and round off upto 2 decimal#
print("Mean",mean)
```

```python
pca = PCA()
MD_pca = pca.fit_transform(mdx)# Take MD_pca is the PCA and
explained_variance = pca.explained_variance_ratio_ # Variance Ratio of each principal component#
print(np.round(explained_variance, 2))#prinitig variance ratio
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(MD_pca[:, 0], MD_pca[:, 1], color='yellow', alpha=0.7)
plt.title("PCA Projection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
```

```python

for i, vector in enumerate(MD_pca.T):
    plt.arrow(0, 0, vector[0]*max(MD_pca[:, 0]), vector[1]*max(MD_pca[:, 1]),
              color='red', alpha=0.7, head_width=0.02, length_includes_head=True)
    plt.text(vector[0]*max(MD_pca[:, 0]) * 1.1,
             vector[1]*max(MD_pca[:, 1]) * 1.1,
             f"Var{i+1}", color='black')

plt.show()
```

```python
from sklearn.cluster import KMeans

# Compute WCSS for different numbers of clusters
wcss = []
max_clusters = 10  # Adjust as needed
for i in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(MD_pca)
    wcss.append(kmeans.inertia_)  # Inertia is the WCSS
```

```python
# Perform k-means clustering on the PCA-transformed data
kmeans = KMeans(n_clusters=3, random_state=42)  # Change n_clusters as needed
kmeans_labels = kmeans.fit_predict(MD_pca)

# Scatter plot of the first two principal components with k-means cluster labels
plt.figure(figsize=(8, 6))
plt.scatter(MD_pca[:, 0], MD_pca[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='X', s=200, label='Centroids')
plt.title("K-Means Clustering on PCA Components")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)
plt.show()
```
